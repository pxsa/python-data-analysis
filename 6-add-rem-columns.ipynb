```python
# Drop the 'url' column (NOT inplace)
tweets.drop(columns='url')
```

```python
# Drop the row with the id of 361388562
# tweets.filter(items=[361388562], axis=0)
tweets[tweets.index == 361388562]
tweets.drop(index=361388562)
```

```python
# Add a new column called 'user'. Set it to 'Joe Biden' for all rows.
tweets['user'] = 'Joe Biden'
tweets
```

```python
# Create a new column called 'ratio'. It should contain the result of dividing each tweet's replies by its retweets
tweets.insert(2, 'ratio', round(tweets['replies'] / tweets['retweets'], 2))
```

```python
tweets.sort_values('ratio', ascending=False)
```

```python
type(tweets['tweet'].str.len())
```

```python
tweets.insert(5, 'length', tweets['tweet'].str.len())
```

```python
tweets
```

```python
# Find the top 10 highest ratio tweets.
tweets.sort_values('ratio').nlargest(10, 'ratio')
```

```python
# Create a column called 'interactions' that contains the sum of each tweet's replies, retweets, quotes, and likes
tweets['interactions'] = tweets['likes'] + tweets['replies'] + tweets['retweets'] + tweets['quotes']
```

```python
# Find the top 8 tweets by number of 'interactions'
tweets.sort_values('interactions', ascending=False).head(8)
```
